### A la recherche de bon 'Tuning' de SVM

```python
%%time

context = zmq.Context()
socket = context.socket(zmq.REP)
socket.bind("tcp://*:5562")


model = SVC(kernel = 'rbf',probability=True,class_weight = 'balanced')
#model = GaussianProcessRegressor(kernel = None)
nbBatch = 10
nbScene = 20
nbIterPerScene = 5
max_bckgrnd_straight = 5

acc_bckgrnd = 0

certitude_obj = 0.8
certitude_bckg = 0.8


batch_y_train_acc_exp = []
batch_y_train_acc_true = []
batch_y_train_acc_pred = []
batch_X_train_acc_rough = []


selected_FPFH_number = 0
#labels = []
#XYZ = np.zeros((424*512*3,3))

#Sending NbBatch, NbScene & NbIterPerScene
message = socket.recv()
send_string = str(nbBatch)
socket.send_string(send_string)
message = socket.recv()
send_string = str(nbScene)
socket.send_string(send_string)
message = socket.recv()
send_string = str(nbIterPerScene)
socket.send_string(send_string)
print("data sent to C++")

for l in range (nbBatch):
    
    X_train_acc_rough = []
    y_train_acc_true = []
    y_train_acc_exp = []
    y_train_acc_pred = []
    print("batch Numero ",l+1)
    
    for i in range(nbScene):
    
        #print("scene Numero ",i+1)
    
        for j in range(nbIterPerScene+1):
            if(j!=0):
                old_y_train_true = np.copy(y_train_true)
                old_y_train_exp = np.copy(y_train_exp)
                old_X_train_rough = np.copy(X_train_rough)
                old_selected_FPFH_number = selected_FPFH_number
            message = socket.recv()
            message1 = array.array('B',message)
            label, FPFHL_preshape = message1[0], message1[1:]
            #labels.append(label)
            FPFH_LAB_L = np.reshape(FPFHL_preshape,(int(len(FPFHL_preshape)/37),37)).astype(int)
            LAB = np.zeros((len(FPFH_LAB_L),3))
            X_train_rough = np.zeros((len(FPFH_LAB_L),36))
            y_train_true = np.zeros(len(FPFH_LAB_L))
            for k in range(len(FPFH_LAB_L)):
                LAB[k] = myfct.rgb2lab(FPFH_LAB_L[k][33:36])
                #On normalise entre 0 et 100
                LAB[k][1] = abs((LAB[k][1]+128)/2.56)
                LAB[k][2] = abs((LAB[k][2]+128)/2.56)
            for k in range(len(FPFH_LAB_L)):
                X_train_rough[k][:33] = FPFH_LAB_L[k][:33]
                X_train_rough[k][33:36] = LAB[k]
                y_train_true[k] = FPFH_LAB_L[k][-1]
            y_train_exp = np.zeros(len(y_train_true))#pas ouf mais permet d'avoir un résultat
            

            if(j!=0):
                compromised_labels = myfct.compromising_data(old_y_train_true,certitude_obj,certitude_bckg)
                label = compromised_labels[old_selected_FPFH_number]
                if(label == 0):
                    acc_bckgrnd += 1
                    
                else:
                    acc_bckgrnd = 0
            
            #if(label == 0):
            #    acc_bckgrnd += 1
                
            #elif(label == 255):#implique 1ere itération du nouvel environnement donc on ne compare pas
            #    pass
            #else:
            #    acc_bckgrnd = 0
                
            #On veux au moins trois FPFH d'objet pour partir sur un modele relativement stable           
            if(sum(y_train_acc_exp) < 3 or sum(y_train_acc_exp)==len(y_train_acc_exp)):
                if(j==0):
                    pass
                else:
                    if(acc_bckgrnd<max_bckgrnd_straight):
                        y_train_acc_exp.append(label)
                        
                        #y_train_acc_exp.append(compromised_labels[old_selected_FPFH_number])
                        y_train_acc_true.append(old_y_train_true[old_selected_FPFH_number])
                        y_train_acc_pred.append(old_y_train_exp[old_selected_FPFH_number])
                        X_train_acc_rough.append(old_X_train_rough[old_selected_FPFH_number])
            
                selected_FPFH_number = int(random.uniform(0,len(X_train_rough)))
            
            else:
                if(j==0):
                    pass
                else:
                    if(acc_bckgrnd<max_bckgrnd_straight):
                        y_train_acc_exp.append(label)
                        #y_train_acc_exp.append(compromised_labels[old_selected_FPFH_number])
                        y_train_acc_true.append(old_y_train_true[old_selected_FPFH_number])
                        y_train_acc_pred.append(old_y_train_exp[old_selected_FPFH_number])
                        X_train_acc_rough.append(old_X_train_rough[old_selected_FPFH_number])
            
                #On réentraine le model avec toutes les données accumulées jusqu'a présent
                X_train_acc = myfct.normalize(X_train_acc_rough)
                model.fit(X_train_acc,y_train_acc_exp)
            
                #
                X_train = myfct.normalize(X_train_rough)
                preds_proba = model.predict_proba(X_train)
                y_train_exp = model.predict(X_train)
                
            
                #selected_FPFH_number = myfct.getUncertainData(X_train_rough,y_train_exp,preds_proba,y_train_acc_exp,1)
                selected_FPFH_number = myfct.getRelevantFPFH(preds_proba,y_train_acc_exp)
            
            
    
            socket.send_string(str(selected_FPFH_number))
            if(label!=255):        
                print("acc_bckgrnd : ",acc_bckgrnd," label : ",label," y_train_true",old_y_train_true[old_selected_FPFH_number])
    batch_y_train_acc_exp.append(y_train_acc_exp)
    batch_y_train_acc_true.append(y_train_acc_true)
    batch_y_train_acc_pred.append(y_train_acc_pred)
    batch_X_train_acc_rough.append(X_train_acc_rough)

    ###DISPLAY FROM MODEL AND NOT REALITY!!!!!!!!!!!!!!!!
mydisplay.displayNormalizedResults(batch_y_train_acc_exp,batch_y_train_acc_true,batch_X_train_acc_rough)
mydisplay.displayNormalizedResults(batch_y_train_acc_pred,batch_y_train_acc_true,batch_X_train_acc_rough)
```

```python


threshold = 0.4

print("différents taux d'erreur pour les samples les moins certains")
is_false, is_true = np.zeros(len(batch_X_train_acc_rough)),np.zeros(len(batch_X_train_acc_rough))
taux_err = np.zeros(len(batch_X_train_acc_rough))
for j in range(len(batch_X_train_acc_rough)):
    nb_samples = len(batch_X_train_acc_rough[j])
    substi_list = np.ones(nb_samples)
    index_list = np.zeros(nb_samples,dtype = int)
    model.fit(myfct.normalize(batch_X_train_acc_rough[j]),batch_y_train_acc_exp[j])
    preds_proba = model.predict_proba(myfct.normalize(batch_X_train_acc_rough[j]))
    for i in range(len(batch_X_train_acc_rough[j])):
        #if(min(preds_proba[i][0],preds_proba[i][1])>threshold):
            #print("preds_proba : ",preds_proba[i])
            #print("y_train_acc_true : ",batch_y_train_acc_true[numBatch][i])
            #print("batch_y_train_acc_pred : ",batch_y_train_acc_pred[numBatch][i])
        #    if(batch_y_train_acc_true[j][i]!=batch_y_train_acc_pred[j][i]):
        #        is_false[j] +=1
        #    else:
        #        is_true[j] +=1
        #for s in range(len(substi_list)):
        if(max(preds_proba[i][0],preds_proba[i][1])<substi_list[substi_list.argmax()]):
            substi_list[substi_list.argmax()] = max(preds_proba[i][0],preds_proba[i][1])
            index_list[substi_list.argmax()] = i
            #break
            
    for s in range(len(substi_list)):
        if(batch_y_train_acc_true[j][index_list[s]]!=batch_y_train_acc_pred[j][index_list[s]]):
            is_false[j] +=1
        else:
            is_true[j] +=1
                
    taux_err[j] = is_false[j]/(is_false[j]+is_true[j])
    print("taux d'erreur : ",int(taux_err[j]*100)/100, "   nb_sample concernés :",int(is_false[j]+is_true[j]))
```

```python
def displayNormalizedResults(batch_y_train_acc_exp,batch_y_train_acc_true,batch_X_train_acc_rough):
    min_len = len(batch_y_train_acc_exp[0])
    for i in range(len(batch_y_train_acc_exp)):
        if(len(batch_y_train_acc_exp[i])<min_len):
            min_len = len(batch_y_train_acc_exp[i])

    true_positive = np.ones(len(batch_y_train_acc_exp))
    true_negative = np.ones(len(batch_y_train_acc_exp))
    false_positive = np.ones(len(batch_y_train_acc_exp))
    false_negative = np.ones(len(batch_y_train_acc_exp))
    precisions = np.zeros((min_len,len(batch_y_train_acc_exp)))
    accuracies = np.zeros((min_len,len(batch_y_train_acc_exp)))
    recalls = np.zeros((min_len,len(batch_y_train_acc_exp)))
    
    

    for j in range(min_len):
        for i in range(len(batch_y_train_acc_exp)):
            if(batch_y_train_acc_exp[i][j] == batch_y_train_acc_true[i][j]):
                if(batch_y_train_acc_exp[i][j] ==1):
                    true_positive[i] += 1
                else:
                    true_negative[i] += 1
            else:
                if(batch_y_train_acc_exp[i][j] ==1):
                    false_positive[i] += 1
                else:
                    false_negative[i] += 1
            precisions[j][i] = true_positive[i]/(true_positive[i]+false_positive[i])
            accuracies[j][i] = (true_positive[i]+true_negative[i])/(true_positive[i]+true_negative[i]+false_positive[i]+false_negative[i])
            recalls[j][i] = true_positive[i]/(true_positive[i]+false_negative[i])
    
    print("err_rate object : ",false_negative/(false_negative+true_positive))
    print("err_rate background : ",false_positive/(false_positive+true_negative))
    X_axis = np.arange(min_len)   
    title = "8 Batchs of 20*5 interaction, T_error from simulation"
    title = "errors due to effector, T_error from simulation"
    y_lim = (0,1)
    x_label = "Number of interactions"

    prec_rec_accu = []
    curves_Names = []
    prec_rec_accu.append(precisions)
    curves_Names.append("precision")
    prec_rec_accu.append(recalls)
    curves_Names.append("recall")
    prec_rec_accu.append(accuracies)
    curves_Names.append("accuracy")
    mydisplay.compare_error_rates_display_once2(prec_rec_accu,X_axis,title,y_lim,x_label,curves_Names)
    
```

```python
batch_y_train_acc_exp_cut = []
batch_y_train_acc_true_cut = []
batch_y_train_acc_pred_cut = []
batch_X_train_acc_rough_cut = []
for i in range(len(batch_y_train_acc_true)):
    if(i!=11):
        batch_y_train_acc_exp_cut.append(batch_y_train_acc_exp[i])
        batch_y_train_acc_true_cut.append(batch_y_train_acc_true[i])
        batch_y_train_acc_pred_cut.append(batch_y_train_acc_pred[i])
        batch_X_train_acc_rough_cut.append(batch_X_train_acc_rough[i])

displayNormalizedResults(batch_y_train_acc_exp_cut,batch_y_train_acc_true_cut,batch_X_train_acc_rough_cut)
displayNormalizedResults(batch_y_train_acc_pred_cut,batch_y_train_acc_true_cut,batch_X_train_acc_rough_cut)
```
